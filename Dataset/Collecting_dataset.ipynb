```python
import numpy as np
import pandas as pd
from datasets import load_dataset
```

```python
# Loading the separate splits of the dataset has train/validation/test
train_dataset = load_dataset("google/code_x_glue_cc_code_refinement", 'medium', split="train")
valid_dataset = load_dataset("google/code_x_glue_cc_code_refinement", 'medium', split="validation")
test_dataset  = load_dataset("google/code_x_glue_cc_code_refinement", 'medium', split="test")
```

```python
# Converting datasets to the pandas
train_df = train_dataset.to_pandas()
valid_df = valid_dataset.to_pandas()
test_df = test_dataset.to_pandas()
```

```python
# Deleting all 'id' columns
del train_df["id"]
del valid_df["id"]
del test_df["id"]
```

```python
# Checking datasets whether contains null value or not
print(train_df.isnull().values.any())
print(valid_df.isnull().values.any())
print(test_df.isnull().values.any())
```

```python
# Saving datasets as csv file
path = r"/content/drive/MyDrive/Bachelor's project"

train_df.to_csv(f'{path}/train.csv')
valid_df.to_csv(f'{path}/validation.csv')
test_df.to_csv(f'{path}/test.csv')
```
