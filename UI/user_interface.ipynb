```python
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer, TextStreamer
from unsloth import FastLanguageModel
```

# Loading the Model

```python
# Loading the Fine-Tuned model
max_seq_length = 2048
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
        model_name = "navidadkhah/Fine-tuned-Llama-3.1-8B-bug-fixing",
        max_seq_length = max_seq_length,
        dtype = dtype,
        load_in_4bit = load_in_4bit,
)
FastLanguageModel.for_inference(model)
```

# Setting up attributes for User Interface

```python
# Setting Up a template for prompting
alpaca_prompt = """{0}\n\n{1}\n\n{2}"""
```

```python
# This function is used for analyzing code and separating the result
def analyze_code(input_code):
    try:
        prompt = alpaca_prompt.format(
            "The following code contains a bug. First, analyze the code and add comments explaining it. Then, identify the bug, fix it, and finally remove the comments. Please do this without any further explanation.",
            input_code,
            ""
        )

        inputs = tokenizer([prompt], return_tensors="pt").to("cuda")

        outputs = model.generate(**inputs, max_new_tokens=512)
        decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract the part after "### Response:" if it exists
        response_start = decoded_output.find("### Response:")
        if response_start != -1:
            return decoded_output[response_start + len("### Response:"):].strip()

        return decoded_output
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
def gradio_interface():
    with gr.Blocks(theme="huggingface", css="""
        #title, #instruction {
            text-align: center;
        }
        button {
            background-color: #add8e6;  # Light blue background color for the button
            color: black;
            display: block;
            margin: 20px auto;  # Adds margin above and below the button and centers it
            padding: 10px 20px;  # Padding around the button text
        }
        """) as demo:
        gr.Markdown("## Bug Fixing Code Analyzer", elem_id="title")
        gr.Markdown("Paste just your code below to analyze and fix bugs:", elem_id="instruction")

        code_input = gr.Textbox(
            label="Input Code", placeholder="Enter your code here...", lines=15
        )

        analyze_button = gr.Button("Analyze and Fix")

        result_output = gr.Textbox(
            label="Output", placeholder="Model output will appear here...", lines=15
        )

        analyze_button.click(
            fn=analyze_code, inputs=code_input, outputs=result_output
        )

    return demo
```

```python
# Launch Gradio app
demo = gradio_interface()
demo.launch()
```
